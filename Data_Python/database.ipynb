```python
import pandas as pd
```

```python
csv = 'Data_bySQL.csv'

data= pd.read_csv(csv)
data.head()
```

```python
data.columns
```

```python
data = data.dropna(subset=['winery', 'wine', 'year', 'rating', 'num_reviews', 'country', 'region', 'price', 'type', 'body', 'acidity', 'designation'])
data = data.reset_index(drop=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Tabla Wineries
wineries = data[['winery', 'country', 'region']].drop_duplicates().reset_index(drop=True)
wineries['id_winery'] = wineries.index + 1  # Generar IDs únicos
wineries.head()
```

```python
types = data[['type']].drop_duplicates().reset_index(drop=True)
types['id_type'] = types.index + 1  # Generar IDs únicos
types.head()
```

```python
designations = data[['designation']].drop_duplicates().reset_index(drop=True)
designations['id_designation'] = designations.index + 1  # Generar IDs únicos
data.head()
```

```python
# Tabla Wines
wines = data.merge(wineries, on=['winery', 'country', 'region'], how='left')
wines = wines.merge(types, on='type', how='left')
wines = wines.merge(designations, on='designation', how='left')
wines = wines[['wine', 'year', 'id_winery', 'id_type', 'id_designation']].drop_duplicates().reset_index(drop=True)
wines['id_wine'] = wines.index + 1  # Generar IDs únicos
wines.head()
```

```python
# Tabla Ratings
ratings = data.merge(wines[['wine', 'id_wine']], on='wine', how='left')
ratings = ratings[['id_wine', 'rating', 'num_reviews', 'price', 'body', 'acidity']].drop_duplicates().reset_index(drop=True)
ratings['id_rating'] = ratings.index + 1  # Generar IDs únicos
ratings.head()
```

```python
# Se crean los archivos CSV

wineries.to_csv("wineries.csv", index=False)
types.to_csv("types.csv", index=False)
wines.to_csv("wines.csv", index=False)
designations.to_csv("designations.csv", index=False)
ratings.to_csv("ratings.csv", index=False) 
```
