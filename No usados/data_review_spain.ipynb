```python
import pandas as pd
```

```python
data = pd.read_csv('winemag-data_2.csv')
data.head()
```

```python
data['country'].value_counts()
```

```python
data_reviews_spain = data[data['country'] == 'Spain']
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.drop('Unnamed: 0', axis=1)
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.drop('description', axis=1)
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.drop('taster_name', axis=1)
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.drop('taster_twitter_handle', axis=1)
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.rename(columns={'points': 'rating'})
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.drop('region_2', axis=1)
data_reviews_spain.head()
```

```python
data_reviews_spain = data_reviews_spain.rename(columns={'region_1': 'region'})
data_reviews_spain.head()
```

```python
data_reviews_spain.to_csv('data_review_spain.csv')
```

```python
data_reviews_spain ['title'] = data_reviews_spain['title'].str.strip().str.lower()
data_reviews_spain.head()
```

```python
data_reviews_spain ['title'] = data_reviews_spain['title'].str.replace(r'[^\w\s]', '', regex=True)
data_reviews_spain.head()
```

```python
palabras_irrelevantes = ['spain', 'vino']

data_reviews_spain['title'] = data_reviews_spain['title'].apply(lambda x: ' '.join(word for word in x.split() if word not in palabras_irrelevantes))
data_reviews_spain.head()
```

```python
data_reviews_spain.to_csv('data_review_spain.csv')
```
