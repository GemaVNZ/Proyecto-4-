```python
url = r'https://raw.githubusercontent.com/GemaVNZ/Proyecto-4-/refs/heads/main/VinoEspaa.csv'
vino_df = pd.read_csv(url, sep=',')
```

```python
unique_values = vino_df['bodega'].unique()
print(unique_values)
```

```python
vino_df['año'] = vino_df['nombre'].str.extract(r'(\b\d{4}\b)')

vino_df['nombre'] = vino_df['nombre'].str[:-5]
vino_df
```

```python
vino_df.dtypes
```

```python
vino_df['like'] = vino_df['like'].astype(int)
```

```python
vino_df['precio'] = vino_df['precio'].str.replace(',', '.').astype(float)
vino_df['valoracion'] = vino_df['valoracion'].str.replace(',', '.').astype(float)


vino_df.info()
```
