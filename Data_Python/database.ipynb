```python
import pandas as pd
```

```python
csv = 'Data_bySQL.csv'

data= pd.read_csv(csv)
data.head()
```

```python
data.columns
```

```python
# Tabla Wineries
wineries = data[["winery"]].drop_duplicates().reset_index(drop=True)
wineries["id"] = range(1, len(wineries) + 1)
wineries.head()
```

```python
# Tabla Wines

wines = data[["wine", "year", "designation", "type", "body", "acidity", "winery"]].drop_duplicates()
wines = wines.merge(wineries, on="winery").drop(columns=["winery"])
wines["id"] = range(1, len(wines) + 1)
wines.head()
```

```python
# Tabla Reviews
reviews = data[["id", "rating", "num_reviews"]]
reviews = reviews.rename(columns={"id": "wine_id"})
reviews.head()
```

```python
# Tabla Regions
regions = data[["country", "region"]].drop_duplicates().reset_index(drop=True)
regions["id"] = range(1, len(regions) + 1)
regions.head()
```

```python
# Tabla Prices
prices = data[["id", "price", "region"]]
prices = prices.rename(columns={"id": "wine_id"})
prices = prices.merge(regions, on="region")[["wine_id", "id", "price"]]
prices = prices.rename(columns={"id": "region_id"})
prices.head()
```

```python
# Se crean los archivos CSV
wineries.to_csv("wineries.csv", index=False)
wines.to_csv("wines.csv", index=False)
reviews.to_csv("reviews.csv", index=False)
regions.to_csv("regions.csv", index=False)
prices.to_csv("prices.csv", index=False)
```
