```python
import pandas as pd
```

```python
csv = 'Data_bySQL.csv'
data_prueba = pd.read_csv(csv)
```

```python
data_prueba.head()
```

```python
# Crear los CSV para las diferentes tablas

# 1. **Tabla de Vinos (wines)**
wines_df = data_prueba[['wine', 'year', 'rating', 'num_reviews', 'price']].copy()
wines_df.rename(columns={
    'wine': 'wine_name',
    'year': 'year',
    'rating': 'rating',
    'num_reviews': 'num_reviews',
    'price': 'price'
}, inplace=True)
wines_df['wine_id'] = wines_df.index + 1  # Asignar un ID único para cada vino
wines_df.to_csv('wines.csv', index=False)

# 2. **Tabla de Viñedos (wineries)**
wineries_df = data_prueba[['winery', 'country', 'region']].drop_duplicates().copy()
wineries_df.rename(columns={
    'winery': 'winery_name',
    'country': 'country',
    'region': 'region'
}, inplace=True)
wineries_df['winery_id'] = wineries_df.index + 1  # Asignar un ID único para cada bodega
wineries_df.to_csv('wineries.csv', index=False)

# 3. **Tabla de Tipos de Vino (wine_types)**
wine_types_df = data_prueba[['type']].drop_duplicates().copy()
wine_types_df['type_id'] = wine_types_df.index + 1  # Asignar un ID único para cada tipo
wine_types_df.to_csv('wine_types.csv', index=False)

# 4. **Tabla de Cuerpos de Vino (wine_bodies)**
wine_bodies_df = data_prueba[['body']].drop_duplicates().copy()
wine_bodies_df['body_id'] = wine_bodies_df.index + 1  # Asignar un ID único para cada cuerpo
wine_bodies_df.to_csv('wine_bodies.csv', index=False)

# 5. **Tabla de Acidez de Vino (wine_acidities)**
wine_acidities_df = data_prueba[['acidity']].drop_duplicates().copy()
wine_acidities_df['acidity_id'] = wine_acidities_df.index + 1  # Asignar un ID único para cada acidez
wine_acidities_df.to_csv('wine_acidities.csv', index=False)

# 6. **Tabla de Designaciones de Vino (wine_designations)**
wine_designations_df = data_prueba[['designation']].drop_duplicates().copy()
wine_designations_df['designation_id'] = wine_designations_df.index + 1  # Asignar un ID único para cada designación
wine_designations_df.to_csv('wine_designations.csv', index=False)

# 7. **Tabla de Relación entre Vinos y Viñedos (wine_winery)**
wine_winery_df = data_prueba[['wine', 'winery']].copy()
wine_winery_df['wine_id'] = wine_winery_df['wine'].apply(lambda x: wines_df[wines_df['wine_name'] == x]['wine_id'].values[0])
wine_winery_df['winery_id'] = wine_winery_df['winery'].apply(lambda x: wineries_df[wineries_df['winery_name'] == x]['winery_id'].values[0])
wine_winery_df = wine_winery_df[['wine_id', 'winery_id']]
wine_winery_df.to_csv('wine_winery.csv', index=False)

# 8. **Tabla de Relación entre Vinos y Tipos (wine_type_rel)**
wine_type_rel_df = data_prueba[['wine', 'type']].copy()
wine_type_rel_df['wine_id'] = wine_type_rel_df['wine'].apply(lambda x: wines_df[wines_df['wine_name'] == x]['wine_id'].values[0])
wine_type_rel_df['type_id'] = wine_type_rel_df['type'].apply(lambda x: wine_types_df[wine_types_df['type'] == x]['type_id'].values[0] if not wine_types_df[wine_types_df['type'] == x].empty else None
)
wine_type_rel_df = wine_type_rel_df[['wine_id', 'type_id']]
wine_type_rel_df.to_csv('wine_type_rel.csv', index=False)

# 9. **Tabla de Relación entre Vinos y Cuerpos (wine_body_rel)**
wine_body_rel_df = data_prueba[['wine', 'body']].copy()
wine_body_rel_df['wine_id'] = wine_body_rel_df['wine'].apply(lambda x: wines_df[wines_df['wine_name'] == x]['wine_id'].values[0])
wine_body_rel_df['body_id'] = wine_body_rel_df['body'].apply(lambda x: wine_bodies_df[wine_bodies_df['body'] == x]['body_id'].values[0] if not wine_bodies_df[wine_bodies_df['body'] == x].empty else None
)
wine_body_rel_df = wine_body_rel_df[['wine_id', 'body_id']]
wine_body_rel_df.to_csv('wine_body_rel.csv', index=False)

# 10. **Tabla de Relación entre Vinos y Acidez (wine_acidity_rel)**
wine_acidity_rel_df = data_prueba[['wine', 'acidity']].copy()
wine_acidity_rel_df['wine_id'] = wine_acidity_rel_df['wine'].apply(lambda x: wines_df[wines_df['wine_name'] == x]['wine_id'].values[0])
wine_acidity_rel_df['acidity_id'] = wine_acidity_rel_df['acidity'].apply(lambda x: wine_acidities_df[wine_acidities_df['acidity'] == x]['acidity_id'].values[0] if not wine_acidities_df[wine_acidities_df['acidity'] == x].empty else None
)
wine_acidity_rel_df = wine_acidity_rel_df[['wine_id', 'acidity_id']]
wine_acidity_rel_df.to_csv('wine_acidity_rel.csv', index=False)

# 11. **Tabla de Relación entre Vinos y Designaciones (wine_designation_rel)**
wine_designation_rel_df = data_prueba[['wine', 'designation']].copy()
wine_designation_rel_df['wine_id'] = wine_designation_rel_df['wine'].apply(lambda x: wines_df[wines_df['wine_name'] == x]['wine_id'].values[0])
wine_designation_rel_df['designation_id'] = wine_designation_rel_df['designation'].apply(lambda x: wine_designations_df[wine_designations_df['designation'] == x]['designation_id'].values[0])
wine_designation_rel_df = wine_designation_rel_df[['wine_id', 'designation_id']]
wine_designation_rel_df.to_csv('wine_designation_rel.csv', index=False)

print("CSV files created successfully.")
```
