```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/GemaVNZ/Proyecto-4-/refs/heads/main/CSV/wines_quality_SPA.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
df_filas_nulas = df[df.isna().all(axis=1)] # visualizar filas nulas

df = df.dropna(how='all')

df
```

```python
# Eliminar filas duplicadas (manteniendo solo la primera apariciÃ³n)
df_sin_duplicados = df.drop_duplicates()

# Verificar el resultado
df_sin_duplicados
```

```python
df_sin_duplicados.dtypes
```

```python
df_sin_duplicados['designation'] = df_sin_duplicados['wine'] + ' ' + df_sin_duplicados['year'].astype(str)
```

```python
df_sin_duplicados
```

```python
df_sin_duplicados['id'] = range(1, len(df_sin_duplicados) + 1)
```

```python
df_sin_duplicados.to_csv('DataFrameGrande.csv', index=False)
df_sin_duplicados
```
